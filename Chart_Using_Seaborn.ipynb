```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
#Load a sample dataset
```

```python
data= pd.read_csv("heart.csv")
print(data.head())
```

```python
sns.lineplot(x="age", y="trestbps", data=data)
plt.title("Line Plot - Age vs Blood Pressure")
plt.show()
```

```python
sns.barplot(x="age", y="fbs", data=data, palette="mako")
plt.title("Bar Plot - Age Vs Fasting Blood Sugar")
plt.show()
```

```python
sns.scatterplot(x="age", y="trestbps", hue="sex",data=data)
plt.title("Scatter Plot - Sex vs Bloodpressure")
plt.savefig("Scatter.png")
plt.show()
```

```python
sns.histplot(data["age"], bins=20, color="orange")
plt.title("Histogram of Age")
plt.show()
```

```python
sns.boxplot(x="age", y="trestbps", data=data, palette="coolwarm")
plt.title("Box Plot - Bloodpressure by Age ")
plt.show()
```

```python
corr= data.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="YlGnBu")
plt.title("Heatmap - Correlation Matrix")
plt.show()
```

```python
sns.pairplot(data, hue="sex")
plt.show()
```

```python
sns.set_theme(style="darkgrid")
sns.pairplot(data, hue="sex")
plt.show()
```
