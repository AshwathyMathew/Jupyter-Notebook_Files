```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("AirQualityUCI.csv", sep=';')
```

```python
df.head()
```

```python
df.info();
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df = df.drop(['Date', 'Time'], axis=1, )
```

```python
 df.info()
```

```python
df = df.drop(['Unnamed: 15', 'Unnamed: 16'], axis=1)
```

```python
df.info()
```

```python
df["PT08.S1(CO)"] = df["PT08.S1(CO)"].fillna(df["PT08.S1(CO)"].mean())
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df["CO(GT)"] = label_encoder.fit_transform(df["CO(GT)"])
df.head(25)
df["CO(GT)"] = df["CO(GT)"].fillna(df["CO(GT)"].mean())
```

```python
df["NMHC(GT)"] = df["NMHC(GT)"].fillna(df["NMHC(GT)"].mean())
df["C6H6(GT)"] = label_encoder.fit_transform(df["C6H6(GT)"])
df["C6H6(GT)"] = df["C6H6(GT)"].fillna(df["C6H6(GT)"].mean())
df["PT08.S2(NMHC)"] = df["PT08.S2(NMHC)"].fillna(df["PT08.S2(NMHC)"].mean())
df["NOx(GT)"] = df["NOx(GT)"].fillna(df["NOx(GT)"].mean())
df["PT08.S3(NOx)"] = df["PT08.S3(NOx)"].fillna(df["PT08.S3(NOx)"].mean())
df["NO2(GT)"] = df["NO2(GT)"].fillna(df["NO2(GT)"].mean())
df["PT08.S4(NO2)"] = df["PT08.S4(NO2)"].fillna(df["PT08.S4(NO2)"].mean())
df["PT08.S5(O3)"] = df["PT08.S5(O3)"].fillna(df["PT08.S5(O3)"].mean())
df["RH"] = label_encoder.fit_transform(df["RH"])
df["AH"] = label_encoder.fit_transform(df["AH"])
df["T"] = label_encoder.fit_transform(df["T"])
df["T"] = df["T"].fillna(df["T"].mean())
df["RH"] = df["RH"].fillna(df["RH"].mean())
df["AH"] = df["AH"].fillna(df["AH"].mean())
```

```python
df.info()
```

```python
plt.bar(df['CO(GT)'], df['AH'])
plt.xlabel('CO(GT)')
plt.ylabel('AH')
plt.show()
```

```python
df_C6H6GT=df['C6H6(GT)'].value_counts()
plt.pie(df_C6H6GT,autopct='%1.1f%%', startangle=90)
plt.title("C6H6GT")
plt.show()
```

```python
df.hist(figsize=(12, 8), bins=30)
plt.suptitle('Distribution of Air Quality Parameters', fontsize=14)
plt.show()
```

```python
df = pd.read_csv("airdataset1.csv")
# Convert Date column to datetime (if applicable)
df['Date'] = pd.to_datetime(df['Date'])

# Set Date as index
df.set_index('Date', inplace=True)

# Plot Air Quality Trends
plt.figure(figsize=(10, 5))
plt.plot(df.index, df['PM2.5'], label='PM2.5', color='red') 
plt.plot(df.index, df['PM10'], label='PM10', color='blue')
plt.xlabel("Date")
plt.ylabel("Pollutant Level")
plt.title("Air Quality Over Time")
plt.legend()
plt.show()
```
