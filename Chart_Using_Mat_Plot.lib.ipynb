```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
#Load a sample dataset
```

```python
data= pd.read_csv("heart.csv")
print(data.head())
```

```python
sns. lineplot(x="age", y="cp", data=data)
plt.title("Line Plot - Age vs Chest Pain")
plt.show()
```

```python
sns.barplot(x="sex", y="age", data=data)
plt.title("Bar Plot - Between Age and Sex")
plt.show()
```

```python
sns.scatterplot(x="sex", y="age", hue="cp", data=data)
plt.title("Scatter Plot - Between Age and Sex")
plt.show()
```

```python
import numpy as np
data= data["age"]
plt.hist(data, bins=10, color='purple', edgecolor='black')
plt.title("Histogram Exmaple")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.savefig("chart.png")
plt.show()
```
